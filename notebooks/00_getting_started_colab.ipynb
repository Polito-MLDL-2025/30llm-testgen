```python
import getpass
import os
import subprocess
from pathlib import Path

REPO_URL = "https://github.com/<ORG_OR_USER>/<REPO>.git"  # TODO: replace with the actual repository URL
TARGET_DIR = Path("/content/adsp-project")

WORKING_BRANCH = "main"
REMOTE_REPO_NAME = "origin"

if not os.getenv("GITHUB_TOKEN"):
    os.environ["GITHUB_TOKEN"] = getpass.getpass("GitHub access token: ")

if TARGET_DIR.exists():
    print("Repository already present, pulling latest changes...")
    subprocess.run([
        "git",
        "-C",
        str(TARGET_DIR),
        "pull",
        REMOTE_REPO_NAME,
        WORKING_BRANCH
    ], check=True)
else:
    clone_cmd = [
        "git",
        "clone",
        f"https://oauth2:{os.environ['GITHUB_TOKEN']}@github.com/<ORG_OR_USER>/<REPO>.git",
        str(TARGET_DIR)
    ]
    subprocess.run(clone_cmd, check=True)

os.chdir(TARGET_DIR)
subprocess.run(["pip", "install", "-r", "requirements.txt"], check=True)
print("Repository ready. You can now import the `adsp` package below.")
```
